# Parameter Shift Visualization

### Visualize parameter shift trajectories across multiple optimizations

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import subprocess
```

```python
# note that gen_steps.sh needs to have proper permissions
def upd8_params(direc, num_params):
    subprocess.check_output(['./gen_steps.sh', str(direc), str(num_params)])   
```

```python
def update_data(direc, num_params):
    params = {'Atom/sigma/tip4p-M': [], 'Atom/epsilon/tip4p-M': [], 'Atom/charge/tip4p-L': [], 
          'VirtualSite/weight1/HOH-3': [], 'VirtualSite/weight1/HOH-4': []}
    # gen step file
    upd8_params(direc, num_params)
    # read in step file
    with open(direc + '/stepz.dat', 'r') as of:
        lines = of.readlines()
    # pull data from lines
    for line in lines:
        l = line.split()
        if ':' in l:
            params[l[-1]].append(float(l[2]))
    # numpy dem up
    for val in params:
        params[val] = np.array(params[val])
    return params
```

```python
def get_color(n):
    return cm.spectral(n/8.,1)
```

```python
plt_plbls = {'Atom/sigma/tip4p-M': r'$\sigma$', 'Atom/epsilon/tip4p-M': r'$\epsilon$', 'Atom/charge/tip4p-L': r'$q$', 
          'VirtualSite/weight1/HOH-3': r'$vs3$', 'VirtualSite/weight1/HOH-4': r'$vs4$'}
```

```python
import matplotlib as mpl
mpl.rc('text', usetex = True)
mpl.rc('font', family = 'serif')
mpl.rcParams['lines.linewidth'] = 1
mpl.rcParams['lines.markersize'] = 3
```

```python
def plt_params(pd, prel, save=None, movie=False):
    fig = plt.figure()
    for i, pdict in enumerate(pd):
        f, s = pdict[prel[0]], pdict[prel[1]]
        # 2d comparison
        if len(prel) == 2:
            ax = plt.subplot(111)
            ax.spines['right'].set_visible(False)
            ax.spines['top'].set_visible(False)
            ax.yaxis.set_ticks_position('left')
            ax.xaxis.set_ticks_position('bottom')
            ax.plot(f, s, '-o', color=get_color(i), label=outs[i])
        # 3d comparison
        if len(prel) == 3:
            ax = fig.gca(projection='3d')
            t = pdict[prel[2]]
            ax.plot(f, s, t, '-o', color=get_color(i), alpha=.66, label=outs[i])
            ax.set_zlabel(plt_plbls[prel[2]])

    ax.set_xlabel(plt_plbls[prel[0]])
    ax.set_ylabel(plt_plbls[prel[1]])
    plt.setp(ax.get_xticklabels(), fontsize=8)
    plt.setp(ax.get_yticklabels(), fontsize=8)
    
    lgd = ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.title('$\Delta \lambda_i (r_c \mbox{ / \AA})$')
    
    if save != None:
        plt.savefig(save + '.png', dpi=250, bbox_extra_artists=(lgd,), bbox_inches='tight')
    plt.show()
    
    # 3d movie
    if movie:
        for ii in np.arange(270,270+180,5):
            ax.view_init(elev=10., azim=ii)
            plt.savefig('movie_' + save + str(ii) + '.png', dpi=250, \
                        bbox_extra_artists=(lgd,), bbox_inches='tight')
```
