```python
def get_data(filename):
    with open(filename,'r') as f:
        lines = f.readlines()
    f.close()
    return lines

def plt_dat(files, num_phase_pts, tgt, fl, colors, properties, units, labels, iter_ls):
    single = True
    if len(files) == 1:
        lines = get_data(files[0])
        save_pref = files[0].split('.')[0]
        if fl:
            save_pref = save_pref + '_fl'
    else:
        single = False
        data_list = []
        final_indices = []
        keys = {}
        for ndx, i in enumerate(files):
            data_list.append(get_data(i))
            num_iters = len([j for j in data_list[ndx] if properties[0] in j])
            if ndx == 0:
                lst = num_iters-1
            else:
                lst = final_indices[ndx-1]+num_iters
            final_indices.append(lst)
            keys[lst] = i
        lines = data_list[0]
        for i in data_list[1:]:
            lines = np.hstack((lines, i))
        save_pref = 'comp'

    # Parse data
    for p_ndx, prop in enumerate(properties):
        # Get line indices
        line_idxs = [(i, line) for i, line in enumerate(lines) if prop in line]
        exp_data = []
        iteration_data = []

        for i in range(len(line_idxs)):
            start = line_idxs[i][0] + 3
            end = start + num_phase_pts
            raw_data = lines[start:end]
            fields = [elem.split() for elem in raw_data]

            if i == 0:
                for j in range(len(fields)):
                    exp_data.append((float(fields[j][0]), float(fields[j][3])))

            tmp_it_data = []
            for elem in fields:
                tmp_it_data.append((float(elem[0]), float(elem[4]), float(elem[6])))
            iteration_data.append(tmp_it_data)

        # plot exp data
        exp_data = np.array(exp_data)
        
        ax = plt.subplot(2, 3, p_ndx)
        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.yaxis.set_ticks_position('left')
        ax.xaxis.set_ticks_position('bottom')
        
        ax.plot(exp_data[:,0], exp_data[:,1], 'o', color='k', label='exp')
            
        # plot else
        iteration_data = np.array(iteration_data)
        for i in [13,0]:#range(len(iteration_data)):
            data = iteration_data[i]
            
            if i in iter_ls:
                ax.errorbar(data[:,0], data[:,1], yerr=data[:,2], fmt=specialmarker[i], 
                    color=special[i], label=speciallbl[i], alpha=specialalpha[i])
            else:
                continue 
        
        # aesthetix
        plt.xlim([225, 475])
        if prop == 'Liquid-H2-RTA Density':
            plt.ylim([850, 1050])
        if prop == 'Liquid-H2-RTA Enthalpy of Vaporization':
            plt.ylim([30, 50])
        plt.title(labels[prop] + ' / ' + units[prop])
        if p_ndx == 5:
            plt.xlabel('$T$ / $K$')
    #lgd = ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    #plt.legend(ncol=4, loc='lower center', bbox_to_anchor=[0.5, 0.5])
    plt.tight_layout()
    plt.savefig('all/p6_0.png', dpi=600)#, bbox_extra_artists=(lgd,), bbox_inches='tight')
    plt.show()
```

```python
def plt_wrap(npp, tgt, f, fl, iter_ls):
    # For plots. -1 is for experimental points.
    # Add support for more colors?
    colors = {-1:'black'} #, 0:'b', 1:'g', 2:'c', 3:'m', 4:'y', 5:'brown', 6:'r'}

    properties = [tgt + ' Density',
                  tgt + ' Enthalpy of Vaporization',
                  tgt + ' Thermal Expansion Coefficient',
                  tgt + ' Isothermal Compressibility',
                  tgt + ' Isobaric Heat Capacity',
                  tgt + ' Dielectric Constant']
    
    lbls = {tgt + ' Density': '$\\rho$',
                  tgt + ' Enthalpy of Vaporization': '$h_{vap}$',
                  tgt + ' Thermal Expansion Coefficient': '$\\alpha_T$',
                  tgt + ' Isothermal Compressibility': '$\\kappa_T$',
                  tgt + ' Isobaric Heat Capacity': '$c_v$',
                  tgt + ' Dielectric Constant': '$\epsilon$'}

    units = {tgt + ' Density': 'kg m$^{-3}$',
             tgt + ' Enthalpy of Vaporization': 'kJ mol$^{-1}$',
             tgt + ' Thermal Expansion Coefficient': '$10^{-4}$ K$^{-1}$',
             tgt + ' Isothermal Compressibility': '$10^{-6}$ bar$^{-1}$',
             tgt + ' Isobaric Heat Capacity': 'cal mol$^{-1}$ K$^{-1}$',
             tgt + ' Dielectric Constant': ' '}

    plt_dat(f, npp, tgt, fl, colors, properties, units, lbls, iter_ls)
```
